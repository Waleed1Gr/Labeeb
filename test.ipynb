```python
import pandas as pd

df = pd.read_csv("data/saudi_human_task_dataset.csv")
df["Name"] = df["Name"].str.replace("الأستاذة", "")
df["Name"] = df["Name"].str.replace("الأستاذ", "")
df["Name"] = df["Name"].str.replace("الآنسة", "")
df["Name"] = df["Name"].str.replace("المهندسة", "")
df["Name"] = df["Name"].str.replace("المهندس", "")
df["Name"] = df["Name"].str.replace("الدكتورة", "")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 6)
# set dark style
sns.set_style("darkgrid")

sns.scatterplot(x=df["Tasks_Assigned"], y=df["Completion_Rate"])
plt.title("Task Completion Rate vs Tasks Assigned")
plt.xlabel("Tasks Assigned")
plt.ylabel("Task Completion Rate")
plt.show()
```
